```python
## First I want to sort the whole data based on the count and take the top five
df1 = df.sort_values('Count', ascending = False).reset_index(drop = True)
df1
```

```python
df1 = df[['Location','Count']].sort_values(by = 'Count',ascending = False).head()
df1
```

```python
df1.plot(x = "Location", y = "Count", kind = "pie", labels = df1.Location, autopct = '%1.2f%%')
plt.legend().set_visible(False)
```

```python
df2 = df.groupby('Region')['Count'].sum().sort_values(ascending = False)
df2
```

```python
df2.plot(kind = 'bar')
plt.show()
```

```python
df3 = df.groupby('Subregion')['Count'].sum().sort_values(ascending = False)
df3
                 
```

```python
df.Year.value_counts()
```

```python
df4 = df[df['Region'].isin(['Asia', 'Europe'])]
df4
```

```python
df4 = df4[df4['Year']>2016][['Region', 'Year', 'Count']]
df4
```

```python
df4 = df4.groupby(['Region', 'Year']).sum()['Count']
df4
```

```python
## To be able to plot the above dataset, I have tounstack it: Meaning I have to convert the data into rows and columns
df_unstacked = df4.unstack(level = 0)
df_unstacked
```

```python
df_unstacked.index = df_unstacked.index.astype(int).astype(str) 
```

```python
df_unstacked.plot(kind = 'line', figsize = (10,5))
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Count of Asia and Europe over years')
plt.show()
```

```python
df7 = df.groupby('Subregion')['Count'].mean().sort_values(ascending = False).round(2)
df7
```

```python
df7.index
```

```python
data = {
    'Category':df7.index,
    'Values' : df7.values,
    'Info' : df7.values 
}
df = pd.DataFrame(data)
df
```

```python
fig = px.treemap(df, path = ['Category'], values = 'Values', title = 'Treemap')
fig.update_traces(hovertemplate = 'Category: %{label}<br>Values:%{value}')
fig.show()
```
